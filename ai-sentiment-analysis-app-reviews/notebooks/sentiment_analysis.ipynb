````python
def  escreve_texto_corretamente(texto):
  texto_novo = texto.upper()
  lista_de_palavras = texto_novo.split()
  nome_final = " ".join(lista_de_palavras)
  return nome_final
````

````python
import pandas as pd
import numpy as np
import random

# Gerar 50 linhas para cada coluna
num_rows = 50

# Lista de exemplos de nomes de produtos
product_names = [
    "Smartphone X", "Laptop Ultra", "Fone de Ouvido Pro", "Smartwatch Vision", "Tablet Go",
    "Câmera Mirrorless", "Drone Explorer", "Caixa de Som Portátil", "Monitor Gamer", "Teclado Mecânico",
    "Mouse Sem Fio", "Roteador Wi-Fi 6", "Webcam Full HD", "Impressora Multifuncional", "Disco SSD 1TB",
    "Memória RAM 16GB", "Processador Intel i7", "Placa de Vídeo RTX", "Console de Jogos", "Controle Gamer"
]

# Lista de categorias de produtos
product_categories = [
    "Eletrônicos", "Informática", "Áudio", "Vestuário", "Casa e Cozinha",
    "Beleza e Cuidados Pessoais", "Esportes", "Brinquedos", "Livros", "Ferramentas"
]

# Gerar dados aleatórios
nomes_produtos = random.choices(product_names, k=num_rows)
categorias_produtos = random.choices(product_categories, k=num_rows)
precos_produtos = np.round(np.random.uniform(50.00, 2500.00, num_rows), 2)
itens_vendidos = np.random.randint(1, 10000, num_rows)
avaliacoes_produtos = np.round(np.random.uniform(1.0, 5.0, num_rows), 1)

# Criar o DataFrame
data = {
    'nome do produto': nomes_produtos,
    'categoria do produto': categorias_produtos,
    'preco do produto': precos_produtos,
    'quantos itens do produto foram vendidos': itens_vendidos,
    'avaliacao do produto': avaliacoes_produtos
}

df_produtos = pd.DataFrame(data)

# Exibir as primeiras 5 linhas do DataFrame
display(df_produtos.head())
````
